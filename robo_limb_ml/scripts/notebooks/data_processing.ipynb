# Use this Notebook to generate training and testing data from cleaned data

```python
import pandas as pd
import numpy as np
import os
from robo_limb_ml.utils.utils import concat_data
```

Getting velocities

```python
train_directory = '../../raw_data/blue_limb_no_cool_down/train'
test_directory = '../../raw_data/blue_limb_no_cool_down/test'
col_order = ['time_begin',
             'time_begin_traj',
             'theta_x',
             'theta_y',
             'vel_x',
             'vel_y',
             'X_throttle',
             'Y_throttle'
             ]
train_data = (concat_data(train_directory, vel=True)
              .drop(columns=['time', 'PX_raw', 'NX_raw', 'PY_raw', 'NY_raw']))[col_order]
test_data = (concat_data(test_directory, vel=True)
             .drop(columns=['time', 'PX_raw', 'NX_raw', 'PY_raw', 'NY_raw']))[col_order]
```

```python
train_data.to_csv('../../ml_data/blue_no_cool_down_train_data.csv', index=False)
test_data.to_csv('../../ml_data/blue_no_cool_down_test_data.csv', index=False)
```
