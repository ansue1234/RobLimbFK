# Exploration algo test

```python
class Cell:
    def __init__(self, x, y):
        self.x = x
        self.y = y
    def set_x(self, x):
        self.x = x
    def set_y(self, y):
        self.y = y

class Coord:
    def __init__(self, x, y):
        self.x = x
        self.y = y
    def set_x(self, x):
        self.x = x
    def set_y(self, y):
        self.y = y
```

```python
max_state = 110
num_rows = 5
num_cols = 5
count_grid = np.zeros(num_rows * num_cols)
total_counts = 0
running_max = 0


def x_y_to_index(x, y):
    global max_state, num_rows, num_cols
    resolution_y = 2 * max_state / num_rows
    resolution_x = 2 * max_state / num_cols
    row = int((y + max_state) / resolution_y)
    col = int((x + max_state) / resolution_x)
    row = max(0, min(num_rows - 1, row))
    col = max(0, min(num_cols - 1, col))
    return row * num_cols + col

def update_count_grid(x, y):
    global total_counts, running_max, count_grid
    index = x_y_to_index(x, y)
    count_grid[index] += 1
    total_counts += 1
    if count_grid[index] > running_max:
        running_max = count_grid[index]

def index_to_x_y(index):
    resolution_y = 2 * max_state / num_rows
    resolution_x = 2 * max_state / num_cols
    row = index // num_cols
    col = index % num_cols
    x = (2 * col + 1) / 2 * resolution_x - max_state
    y = (2 * row + 1) / 2 * resolution_y - max_state
    return Coord(x, y)

def get_sampled_x_y(sampled_val):
    global total_counts, running_max, count_grid
    counts_tot = (running_max + 1) * num_rows * num_cols - total_counts
    target = sampled_val * counts_tot
    # print("running max: ", running_max)
    # print("total counts: ", total_counts)
    # print("counts_tot: ", counts_tot)
    # print("sampled_val: ", sampled_val)
    # print("target: ", target)
    for i in range(num_rows * num_cols):
        target -= ((running_max + 1) - count_grid[i])
        if target <= 0:
            # print("i", i)
            # print("index_to_x_y(i)", index_to_x_y(i).x, index_to_x_y(i).y)
            return index_to_x_y(i)

def x_y_to_cell(x, y):
    global max_state, num_rows, num_cols
    resolution_y = 2 * max_state / num_rows
    resolution_x = 2 * max_state / num_cols
    row = int((y + max_state) / resolution_y)
    col = int((x + max_state) / resolution_x)
    row = max(0, min(num_rows - 1, row))
    col = max(0, min(num_cols - 1, col))
    # print("row: ", row)
    # print("col: ", col)
    return Cell(col, row)
```

```python
from tqdm import tqdm
total_iter = 10000
current_coord = Coord(np.random.uniform(-max_state, max_state), np.random.uniform(-max_state, max_state))
coords_x = []
coords_y = []
grids = np.zeros(num_rows*num_cols)
for _ in tqdm(range(total_iter)):
    rand_select_thresh = np.random.uniform(0, 1)
    sampled_coord = get_sampled_x_y(rand_select_thresh)


    sampled_cell = x_y_to_cell(sampled_coord.x, sampled_coord.y)
    current_cell = x_y_to_cell(current_coord.x, current_coord.y)
    resolution_y = 2 * max_state / num_rows
    resolution_x = 2 * max_state / num_cols
    x_diff = sampled_cell.x - current_cell.x
    y_diff = sampled_cell.y - current_cell.y

    if x_diff == 0 and y_diff == 0:
        rand_dir = np.random.randint(0, 8)
        if rand_dir == 0:
            x_diff = 1
            y_diff = 1
        elif rand_dir == 1:
            x_diff = 1
            y_diff = 0
        elif rand_dir == 2:
            x_diff = 1
            y_diff = -1
        elif rand_dir == 3:
            x_diff = 0
            y_diff = 1
        elif rand_dir == 4:
            x_diff = 0
            y_diff = -1
        elif rand_dir == 5:
            x_diff = -1
            y_diff = 1
        elif rand_dir == 6:
            x_diff = -1
            y_diff = 0
        else:
            x_diff = -1
            y_diff = -1
    
    if x_diff > 0:
        delta_x = np.random.uniform(0, 0.5*resolution_x)
    elif x_diff == 0:
        delta_x = 0
    else:
        delta_x = -np.random.uniform(0, resolution_x)

    if y_diff > 0:
        delta_y = np.random.uniform(0, 0.5*resolution_y)
    elif y_diff == 0:
        delta_y = 0
    else:
        delta_y = -np.random.uniform(0, resolution_y)
    # print("x_diff: ", x_diff)
    # print("y_diff: ", y_diff)
    # print("resolution_x", resolution_x)
    # print("resolution_y", resolution_y)
    new_x = np.clip(current_coord.x + delta_x, -max_state, max_state)
    new_y = np.clip(current_coord.y + delta_y, -max_state, max_state)
    current_coord.set_x(new_x)
    current_coord.set_y(new_y)
    update_count_grid(current_coord.x, current_coord.y)
    # if _ % 1 == 0:
        # print(rand_select_thresh)
    # print(current_coord.x, current_coord.y)
    coords_x.append(current_coord.x)
    coords_y.append(current_coord.y)
    # print(count_grid)
    if _ % 1000 == 0 and _ != 0:
        count_tot = (running_max + 1)*num_rows*num_cols - total_counts
        prob_grid = (running_max + 1 - count_grid) / count_tot
        grids = np.vstack((grids, prob_grid))
        
```

```python
PWMMaxPX = 30.00
PWMMinPX = 20.00
PWMMaxNX = 35.00
PWMMinNX = 20.00
PWMMaxPY = 34.00
PWMMinPY = 21.00
PWMMaxNY = 30.00
PWMMinNY = 19.00

def PWMRange(pwm_signals, PWM_P_Max, PWM_P_Min, PWM_N_Max, PWM_N_Min):
    num_intervals = len(pwm_signals) // 2
    p_increment = (PWM_P_Max - PWM_P_Min) / (num_intervals - 1)
    n_increment = (PWM_N_Max - PWM_N_Min) / (num_intervals - 1)
    
    # Negatives
    for i in range(num_intervals):
        pwm_signals[i] = (-PWM_N_Max + i * n_increment)*-1
    
    pwm_signals[num_intervals] = 0
    print(pwm_signals)
    # Positives
    for j in range(num_intervals):
        pwm_signals[num_intervals + j + 1] = PWM_P_Min + j * p_increment

# Example usage
num_PWM_signals = 21
pwm_signals = [0] * (num_PWM_signals)
PWMRange(pwm_signals, PWMMaxPX, PWMMinPX, PWMMaxNX, PWMMinNX)
print(pwm_signals)
```
