```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import os
```

```python
# # Define the specific length you want to filter by
color = 'purple'
directory = '../data/raw_data/'+color
for filename in os.scandir(directory):
    if filename.is_file():
        if 'wobble' in filename.name:
            specific_length = 11
        else:
            specific_length = 9
        # Open the file for reading
        with open(filename.path, 'r') as file:
            # Read all lines from the file
            lines = file.readlines()
        print(filename.name, specific_length)
        # Filter the lines based on the specified length
        filtered_lines = [line for line in lines if len(line.strip().split(',')) == specific_length]

        with open('../data/cleaned_data/'+color +'/' + filename.name, 'w') as filtered_file:
            if specific_length == 9:
                first_line = 'time,theta_x,theta_y,X_throttle,Y_throttle,PX_raw,NX_raw,PY_raw,NY_raw\n'
            else:
                first_line = 'time,theta_x,theta_y,pwm_px,pwm_nx,pwm_py,pwm_ny,pwm_px_raw,pwm_nx_raw,pwm_py_raw,pwm_ny_raw\n'
            filtered_file.write(first_line)
            for line in filtered_lines:
                filtered_file.write(line)
```

```python
color = 'purple'
directory = '../data/cleaned_data/'+color
for filename in os.scandir(directory):
    # print(filename.path)
    data = pd.read_csv(str(filename.path))
    plt.figure()
    plt.plot(data['time']/1000, data['theta_x'], label='theta_x')
    plt.plot(data['time']/1000, data['theta_y'], label='theta_y')
    plt.legend()
    plt.title(filename.name)
```

```python
colors = ['blue','teal','white', 'purple']
total_samples = 0
for color in colors:
    directory = '../data/cleaned_data/'+color
    color_samples = 0
    for filename in os.scandir(directory):
        # print(filename.path)
        data = pd.read_csv(str(filename.path))
        color_samples += len(data)
    print(color, color_samples)
    total_samples += color_samples
print('Total samples:', total_samples)
```
